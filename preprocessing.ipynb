```python
# for each file in train_negative
for file in os.listdir(os.path.join(data, "train", "neg")):
    # open text file 
    with open(os.path.join(data, "train", "neg", file), "r") as train_file:
        for line in train_file:
            # append content of text file to data frame 
            train_df = train_df.append({"Text": line, "Sentiment": "negative" }, ignore_index=True)
    
```

```python
train_df
```

```python
train_df = train_df.sample(
    frac=1,
    random_state=1
).reset_index()
```

```python
train_df = train_df.drop(columns="index")
```

```python
train_df
```

```python
# convert to_csv file 
path = "/home/maha.agro/thesis/bert_lnl/BERT-LNL/data"
train_df.to_csv(os.path.join(path, "IMDB", "train.csv"), index=False)
```
